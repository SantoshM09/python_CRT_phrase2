```python
#Prime number
a=int(input("enter a number:"))
for i in range(2,a):
  if a%i==0:
    print(a,"is not a prime number")
    break
else:
  print(a,"is a prime number")
```

```python
#Prime number within a given range
a=int(input("enter a number:"))
b=int(input("enter a number:"))
for i in range(a,b+1):
  if i>1:
    for j in range(2,i):
      if i%j==0:
        break
    else:
      print(i)
```

```python
#find out the second max element from the list
a=[1,5,3,4,2]
first_max=a[0]
second_max=a[0]
for i in a:
  if i>first_max:
    second_max=first_max
    first_max=i
  elif i>second_max:
    second_max=i
print(second_max)
```

```python
#find out the second min element from the list
a=[1,5,3,4,2]
first_min=a[0]
second_min=a[0]
for i in a:
  if i<first_min:
    second_max=first_min
    first_min=i
  elif i<second_max:
    second_min=i
print(second_min)
```

```python
#print the occurence of each element in the list
a=[1,5,3,5,1,1,4,2]
dict={}
for i in a:
  if i in dict:
    dict[i]=dict[i]+1
  else:
    dict[i]=1
print(dict)
```

```python
#print all dublicates elements in the list
a=[1,5,3,5,1,1,4,2]
dict={}
for i in a:
  dict[i]=dict.get(i,0)+1
for key,value in dict.items():
  print(key,"--",value)
```
